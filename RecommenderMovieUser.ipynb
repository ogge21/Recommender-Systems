### User Based Recommender System for Movies

```python
import pandas as pd
```

```python
df=pd.read_csv('TMDb_updated.CSV')
```

```python
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'user'},inplace=True)
```

```python
df.head()
```

```python
import numpy as np
```

```python
movieratings=df.pivot_table(index=['user'],columns=['title'],values='vote_average')
```

```python
movieratings.head()
```

```python
Bloodshot=movieratings['Bloodshot']
```

```python
Bloodshot.head()
```

```python
movieratings[['Bloodshot','Ad Astra']].corr()
```

```python
movieratings.corrwith(Bloodshot).sort_values(ascending=False)
```
